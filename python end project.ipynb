```python
#Q2.Segregate the employees w.r.t different positions.
```

```python
position_counts = df['Position'].value_counts()
position_counts 
```

```python
plt.bar(position_counts.index,position_counts.values)
plt.xlabel('Position')
plt.ylabel('Number of employees')
plt.show()
```

```python
#Q3.Find from which age group most of the employees belong to.
```

```python
ag=df["Age"].value_counts().sort_values()
ag
```

```python
age_groups = pd.cut(df['Age'], bins=[20, 25, 30, 35, 40, 45, 50, 55, 60], labels=['20-25', '26-30', '31-35', '36-40', '41-45', '46-50', '51-55', '55+'])
age_group_counts = age_groups.value_counts()
age_group_counts
```

```python
plt.bar(age_group_counts.index,age_group_counts.values)
plt.xlabel('Age group')
plt.ylabel('Number of employees')
plt.show()
```

```python
age_groups = age_group_counts.idxmax()
number=age_group_counts.max()
print("Most of the employees belongs to the age group",age_groups,"with total number of employees",number)
```

```python
#Q4.Find out under which team and position, spending in terms of salary is high.
```

```python
salary_spending = df.groupby(['Team', 'Position'])['Salary'].sum()
salary_spending 
```

```python
high_salary = salary_spending.idxmax()
print('Team and position under which salary is high is', high_salary)
```

```python
#Q5.Find if there is any correlation between age and salary , represent it visually.
```

```python
correlation = df['Age'].corr(df['Salary'])
correlation
```

```python
sns.scatterplot(x='Age',y='Salary',data=df)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Corelation between Age and Salary')
plt.show()
```
