```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file = pd.read_csv('D:/fifa_data.csv')
```

```python
#question 1
```

```python
most_players = file['Nationality'].value_counts().idxmax()
print("Country with the most number of players is", most_players ,"with No. of players", file['Nationality'].value_counts().max())
```

```python
#question 2
```

```python
top5_players = file['Nationality'].value_counts().head(5)
plt.bar(top5_players.index, top5_players.values)

plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.title('Top 5 Countries with Most Number of Players')
plt.show()
```

```python
#question 3
```

```python
file['Wage'] = file['Wage'].str.replace("€","")
file['Wage'] = file['Wage'].str.replace("k","")
file['Wage'] = file['Wage'].str.replace("K","")
file['Wage'] = pd.to_numeric(file['Wage'])
highest_salary = file.loc[file['Wage'].idxmax()]
print("Player with the highest salary:", highest_salary['Name'])
```

```python
#question 4
```

```python
plt.hist(file['Wage'])
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Range of Players')
plt.show()
```

```python
#question 5
```

```python
file['Height'] = file['Height'].str.replace("'",".")
file['Height'] = pd.to_numeric(file['Height'])
tallest_player = file.loc[file['Height'].idxmax()]
print("Tallest player:", tallest_player['Name'])
```

```python
#question 6
```

```python
most_players = file['Club'].value_counts().idxmax()
print("Club with the most number of players:", most_players)
```

```python
#question 7
```

```python
preferred_foot = file['Preferred Foot'].value_counts()
plt.bar(preferred_foot.index, preferred_foot.values)
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.title('Preferred Foot of Players')
plt.show()
```
